**COEFICIENTE DE CORRELACIÓN DE PEARSON**

image.png

```python
import numpy as np

x = np.array([185, 170, 168, 179, 182, 188])
y = np.array([72, 56, 60, 68, 72, 77])

# Cálculo del coeficiente de correlación de Pearson (r)
numerador = np.sum((x - np.mean(x)) * (y - np.mean(y)))
denominador = np.sum((x - np.mean(x))**2) * np.sum((y - np.mean(y))**2)

if denominador > 0:
    r = numerador / np.sqrt(denominador)
else:
    r = 0

print('El coeficiente de correlación de Pearson (r) es:', round(r, 2))
```

**SIGNIFICADO DEL COEFICIENTE DE CORRELACIÓN DE PEARSON**

| Valor r          | Descripción                              |
|------------------|------------------------------------------|
| +0.70 o superior | Relación positiva muy fuerte             |
| +0.40 a +0.69    | Fuerte relación positiva                 |
| +0.30 a +0.39    | Relación positiva moderada               |
| +0.20 a +0.29    | Relación positiva débil                  |
| +0.01 a +0.19    | Ninguna o insignificante relación        |
| 0                | Sin relación [correlación cero]          |
| -0.01 a -0.19    | Ninguna o insignificante relación        |
| -0.20 a -0.29    | Relación negativa débil                  |
| -0.30 a -0.39    | Relación negativa moderada               |
| -0.40 a -0.69    | Fuerte relación negativa                 |
| -0.70 o superior | Relación negativa muy fuerte             |
